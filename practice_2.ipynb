### Инициализация

```python
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf

sp_conf = SparkConf() \
    .setAppNaame('NewProLab Practice #1') \
    .set('spark.executor.memory', '2g') \
    .set('spark.executor.cores', 1) \
    .set('spark.executor.instances', 8)

spark = SparkSession.builder \
    .config(conf=sp_conf) \
    .getOrCreate()
```

### Загружаем данные в память

```python
df = spark.read \
    .load(
        'npl_churn_train.csv',
        format='csv',
        header=True,
        inferSchema=True
    ) \
    .na.fill(0) \
    .repartition(32) \
    .cache()

df.count()
#df.show()
```

### Эксплоративный анализ

```python
df.printSchema()
```

```python
empty_cols = []

for c in df.columns:
    print('=' * 3, c, '=' * 3)
    df.select(c).describe().show()
    
    p = df.select(c).describe().toPandas()
    try:
        if float(p[p['summary'] == 'mean'][c].tolist()[0]) == .0 and \
            float(p[p['summary'] == 'stddev'][c].tolist()[0])== .0:
            empty_cols.append(c)
    except: continue
```

```python
df = df.drop(empty_cols)
```

```python
to_normalize = [
    'CR_PROD_CNT_IL',
    'APP_REGISTR_RGN_CODE',
    'CR_PROD_CNT_VCU',
    'CR_PROD_CNT_TOVR',
    'CR_PROD_CNT_PIL',
    'CR_PROD_CNT_CC',
    'LDEAL_TENOR_MAX',
    'LDEAL_YQZ_CHRG',
    'CR_PROD_CNT_CCFP',
    'DEAL_YQZ_IR_MAX',
    'DEAL_YQZ_IR_MIN',
    'LDEAL_TENOR_MIN',
    'DEAL_GRACE_DAYS_ACC_S1X1',
    'AVG_PCT_MONTH_TO_PCLOSE',
    'DEAL_YWZ_IR_MIN',
    'DEAL_YWZ_IR_MAX',
    'DEAL_GRACE_DAYS_ACC_AVG',
    'MAX_PCLOSE_DATE',
    'LDEAL_YQZ_PC',
    'CLNT_SETUP_TENOR',
    'DEAL_GRACE_DAYS_ACC_MAX',
    'AVG_PCT_DEBT_TO_DEAL_AMT'
    
]

to_bucketize = [
    'REST_AVG_CUR',
    'TURNOVER_CC',
    'TURNOVER_PAYM',
    'AGE',
    'REST_AVG_PAYM',
    'LDEAL_AMT_MONTH',
    'CLNT_SALARY_VALUE'
]

categorical = [
    'CLNT_TRUST_RELATION', 
    'APP_MARITAL_STATUS',
    'CLNT_JOB_POSITION_TYPE',
    'APP_DRIVING_LICENSE',
    'APP_EDUCATION',
    'APP_TRAVEL_PASS',
    'APP_CAR',
    'APP_POSITION_TYPE',
    'APP_EMP_TYPE',
    'APP_COMP_TYPE',
    'APP_KIND_OF_PROP_HABITATION'
]

text = [
    'CLNT_JOB_POSITION'
]
```

```python
other = []
for c in df.columns:
    if c not in to_normalize + to_bucketize + categorical + text + ['_c0', 'TARGET', 'PACK']:
        other.append(c)
```

### Сборка пайплайна

```python
from pyspark.ml.feature import (
    VectorAssembler,
    MinMaxScaler,
    MaxAbsScaler,
    OneHotEncoderEstimator,
    Bucketizer,
    StringIndexer,
    QuantileDiscretizer
)
from pyspark.ml.pipeline import Pipeline
from pyspark.ml.classification import (
    LogisticRegression,
    RandomForestClassifier,
    GBTClassifier
)
```

#### Нормирование

```python
print(VectorAssembler().explainParams())
```

```python
print(MinMaxScaler().explainParams())
```

```python
normalization_stages = [
        VectorAssembler(
            inputCols=to_normalize,
            outputCol='n_features',
        ),
        MinMaxScaler(
            min=0.,
            max=1.,
            inputCol='n_features',
            outputCol='normalized'
        ),
    ]
```

#### Квантилизация

```python
print(QuantileDiscretizer().explainParams())
```

```python
quantilization_stages = [
        QuantileDiscretizer(
            numBuckets=10,  
            handleInvalid='skip',
            inputCol=c,
            outputCol='bucketized_' + c,
            relativeError=0.01
        ) for c in to_bucketize
    ]
```

#### Обработка категориальных признаков

```python
print(StringIndexer().explainParams())
```

```python
print(OneHotEncoderEstimator().explainParams())
```

```python
categorical_stages = [
    StringIndexer(
        inputCol=c,
        outputCol='cat_' + c,
        handleInvalid='skip'
    ) for c in categorical
] + [
    OneHotEncoderEstimator(
        inputCols=['cat_' + c for c in categorical],
        outputCols=['ohe_' + b for b in categorical],
        handleInvalid='keep', 
        dropLast=True
    )
]
```

#### Сборка векторов

```python
print(VectorAssembler().explainParams())
```

```python
vectorization_stages = [
    VectorAssembler(
        inputCols=['normalized'] + 
            ['ohe_' + b for b in categorical] + 
            ['bucketized_' + c for c in to_bucketize] +
            other,
        outputCol='features'
    ),
    VectorAssembler(
        inputCols=to_normalize + 
            ['ohe_' + b for b in categorical] + 
            ['bucketized_' + c for c in to_bucketize] +
            other,
        outputCol='tree_features'
    )
]
```

#### Классификаторы

```python
print(LogisticRegression().explainParams())
```

```python
print(GBTClassifier().explainParams())
```

```python
print(RandomForestClassifier().explainParams())
```

```python
classification_stages = [
    LogisticRegression(
        featuresCol='features', 
        labelCol='TARGET', 
        predictionCol='lr_prediction', 
        maxIter=100, 
        regParam=0.0, 
        elasticNetParam=0.0, 
        tol=1e-06, 
        fitIntercept=True, 
        threshold=0.5, 
        #thresholds=None, 
        probabilityCol='lr_probability', 
        rawPredictionCol='lr_rawPrediction', 
        standardization=True, 
        aggregationDepth=2, 
        family='auto'
    ),
    GBTClassifier(
        featuresCol='tree_features', 
        labelCol='TARGET', 
        predictionCol='gbt_prediction', 
        maxDepth=2, 
        maxBins=10, 
        minInstancesPerNode=3, 
        minInfoGain=0.01, 
        maxMemoryInMB=256, 
        cacheNodeIds=False, 
        checkpointInterval=10, 
        lossType='logistic', 
        maxIter=32, 
        stepSize=0.05, 
        seed=None, 
        subsamplingRate=0.8
    ),
    RandomForestClassifier(
        featuresCol='tree_features', 
        labelCol='TARGET', 
        predictionCol='rf_prediction', 
        probabilityCol='rf_probability', 
        rawPredictionCol='rf_rawPrediction', 
        maxDepth=2, 
        maxBins=10, 
        minInstancesPerNode=1, 
        minInfoGain=0.01, 
        maxMemoryInMB=256, 
        cacheNodeIds=False, 
        checkpointInterval=10, 
        impurity='gini', 
        numTrees=32, 
        featureSubsetStrategy='auto', 
        seed=None, 
        subsamplingRate=0.8
    )
]
```

### Пайплайн

```python
from pyspark.ml.pipeline import Pipeline


newprolab_pipeline = Pipeline(
    stages = normalization_stages +
        quantilization_stages +
        categorical_stages +
        vectorization_stages +
        classification_stages
)
```

```python
train, test = df.randomSplit([0.75, 0.25])
```

```python
PIPE = newprolab_pipeline.fit(train)
```

```python
PIPE.save('my_pipeline')
```

```python
result_on_train = PIPE.transform(train).cache()
result_on_test = PIPE.transform(test).cache()
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

lr_evaluator = BinaryClassificationEvaluator(
        rawPredictionCol='lr_prediction', labelCol='TARGET', metricName='areaUnderROC'
    )

rf_evaluator = BinaryClassificationEvaluator(
        rawPredictionCol='rf_prediction', labelCol='TARGET', metricName='areaUnderROC'
    )

gbt_evaluator = BinaryClassificationEvaluator(
        rawPredictionCol='gbt_prediction', labelCol='TARGET', metricName='areaUnderROC'
    )

print('= Logistic Regression =')
print('ROC AUC on train: %.5f' % lr_evaluator.evaluate(result_on_train))
print('ROC AUC on test: %.5f' % lr_evaluator.evaluate(result_on_test))

print('= Random Forest =')
print('ROC AUC on train: %.5f' % rf_evaluator.evaluate(result_on_train))
print('ROC AUC on test: %.5f' % rf_evaluator.evaluate(result_on_test))

print('= Gradient Boosting =')
print('ROC AUC on train: %.5f' % gbt_evaluator.evaluate(result_on_train))
print('ROC AUC on test: %.5f' % gbt_evaluator.evaluate(result_on_test))
```

### Кросс-валидация

```python
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

grid = ParamGridBuilder() \
    .addGrid(QuantileDiscretizer.numBuckets, [3, 10, 16]) \
    .addGrid(LogisticRegression.regParam, [0.1, 0.01]) \
    .addGrid(LogisticRegression.elasticNetParam, [0.1, 0.01]) \
    .addGrid(GBTClassifier.maxDepth, [2, 5, 7]) \
    .addGrid(GBTClassifier.maxBins, [16, 32]) \
    .addGrid(RandomForestClassifier.maxDepth, [2, 5, 7]) \
    .addGrid(RandomForestClassifier.maxBins, [16, 32]) \
    .build()

print(grid)
```
